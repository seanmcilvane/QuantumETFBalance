```python
from qiskit import Aer
from qiskit.algorithms.optimizers import *
from QAOA import Portfolio_Optimization
```

```python

backend = Aer.get_backend('qasm_simulator')
optimizer = COBYLA(maxiter = 20, disp = True)
Stocks = ['AMZN', 'ARQQ', 'GOOGL', 'IBM', 'IONQ']
```

```python
instance = Portfolio_Optimization(Stocks, optimizer, backend)
```

```python
instance.plot_covariance_matrix()
```

```python
print(instance.classical_solution())
```

```python
qc = instance.layer_circuit()
qc.draw()
```

```python
print(instance.QAOA())
```
